```python
index = df_distrito.state.value_counts().index
ax = sns.countplot(data= df_distrito, y = 'state', order = index, )
plt.title('Estados')
plt.show()
```

Se puede observar una predominancia de los estados de Connecticut, Utah y Massachusetts

```python
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Locale Distribution', size = 20, font="Serif")
explode = (0.05, 0.05, 0.05, 0.05)
labels = list(df_distrito.locale.value_counts().index)
sizes = df_distrito.locale.value_counts().values
ax.pie(sizes, explode=explode,startangle=60, labels=labels,autopct='%1.0f%%', pctdistance=0.7, colors=["#d45d00","#ff9100","#eaaa00","#6d6875"])
ax.add_artist(plt.Circle((0,0),0.4,fc='white'))
plt.show()
```

La mayoria de la gente vive en los suburvios

```python
c1=c2=c3=0
for s in df_producto["Sector(s)"]:
    if(not pd.isnull(s)):
        s = s.split(";")
        for i in range(len(s)):
            sub = s[i].strip()
            if(sub == 'PreK-12'): c1+=1
            if(sub == 'Higher Ed'): c2+=1
            if(sub == 'Corporate'): c3+=1

fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Distribucion por sector', size = 20, font="Serif")
explode = (0.05, 0.05, 0.05)
labels = ['PreK-12','Higher Ed','Corporate']
sizes = [c1,c2, c3]
ax.pie(sizes, explode=explode,startangle=60, labels=labels,autopct='%1.2f%%', pctdistance=0.7, colors=["#ff228a","#20b1fd","#ffb703"])
ax.add_artist(plt.Circle((0,0),0.4,fc='white'))
plt.show()
```

Los productos estan dirigidos principalmente al sector PreK-12

```python
primary_essential_main = []
primary_essential_sub = []
for s in df_producto["Primary Essential Function"]:
    if(not pd.isnull(s)):
        s1 = s.split("-",1)[0].strip()
        primary_essential_main.append(s1)
    else:
        primary_essential_main.append(np.nan)

    if(not pd.isnull(s)):
        s2 = s.split("-",1)[1].strip()
        primary_essential_sub.append(s2)
    else:
        primary_essential_sub.append(np.nan)

df_producto["primary_essential_main"] = primary_essential_main
df_producto["primary_essential_sub"] = primary_essential_sub
```

```python
plt.figure(figsize=(16, 20))
sns.countplot(y='primary_essential_sub', data=df_producto, order=df_producto["primary_essential_sub"].value_counts().index)
plt.title("Primary Essential Function(Sub)",font="Serif", size=20)
plt.show()
```
